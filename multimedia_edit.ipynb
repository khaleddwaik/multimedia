```python


# 2) Imports
from moviepy.editor import VideoFileClip
from moviepy.video.fx.all import speedx, blackwhite
from IPython.display import Video, display
from moviepy.video.fx.all import rotate, resize, crop, mirror_x
# 3) Paths & params
input_path  = "/content/IMG_7091.mp4"
output_path = "/content/last.mp4"
fadein      = 1.0
fadeout     = 1.0

# 4) Display original
display(Video(input_path, embed=True))

# 5) Load & edit
clip = VideoFileClip(input_path)
orig_w, orig_h = clip.size

trim_start, trim_end = 1.0, 16
clip = clip.subclip(trim_start, trim_end)
clip = clip.fadein(fadein).fadeout(fadeout)
clip = clip.resize((orig_w, orig_h))
clip = crop(clip,
            x1=0.1*orig_w, y1=0.1*orig_h,
            x2=0.9*orig_w, y2=0.9*orig_h)

clip.write_videofile(
    output_path,
    codec="libx264",
    audio_codec="aac",
    verbose=False,
    logger=None
)

# 7) Display edited result
display(Video(output_path, embed=True))

```

```python
 # 1) Install pydub
!pip install -q pydub


input_path = "/content/AUDIO-2025-05-17-17-22-53.m4a"  # e.g. "my_voice_note.m4a"

# 3) Imports
from pydub import AudioSegment
from IPython.display import Audio

# 4) Editing parameters â€” adjust as needed
start_sec   = 1    # trim start time (seconds)
end_sec     = 11   # trim end time (seconds)
gain_db     = -5.0   # volume change in dB (positive or negative)
fadein_sec  = 1.5    # fade-in duration (seconds)
fadeout_sec = 1.5    # fade-out duration (seconds)
output_path = 'edited_voice.mp4'

# 5) Load the M4A (explicitly specify format)
audio = AudioSegment.from_file(input_path, format="m4a")

# 6) Trim
audio = audio[int(start_sec*1000) : int(end_sec*1000)]

# 7) Change volume
audio = audio + gain_db

# 8) Apply fade in/out
audio = audio.fade_in(int(fadein_sec*1000)).fade_out(int(fadeout_sec*1000))

# 9) Export as M4A with AAC codec
audio.export(output_path, format="mp4", codec="aac")

# 10) Play it inline
Audio(output_path, autoplay=True)
```

```python
from PIL import Image, ImageEnhance, ImageFilter, ImageOps

# 1. Load your image (replace with your own file path)
img = Image.open('/content/IMG_3683.jpg')

# 1a. Correct orientation based on EXIF data
img = ImageOps.exif_transpose(img)

# 2. Editing Function 1: Increase contrast by 50%
contrast_enhancer = ImageEnhance.Contrast(img)
img_contrast = contrast_enhancer.enhance(1.5)

# 3. Editing Function 2: Apply a Gaussian blur (radius=2)
img_blur = img_contrast.filter(ImageFilter.GaussianBlur(radius=2))

# 4. Convert to grayscale
img_gray = ImageOps.grayscale(img_blur)

# 5. Convert to monochrome (black & white)
img_bw = img_gray.point(lambda x: 0 if x < 128 else 255, mode='1')

# 6. Save all three versions
img_blur.save('edited_color.jpg')
img_gray.save('grayscale.jpg')
img_bw.save('monochrome.png')

print("Saved: edited_color.jpg, grayscale.jpg, monochrome.png")
```
